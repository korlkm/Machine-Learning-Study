```python

import requests
from bs4 import BeautifulSoup

def get_book_title(isbn):
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'

  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')
  title=soup.find('a', attrs={'class':'gd_name'}).get_text()

  return title
```

```python
def get_book_author(isbn):

  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'

  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')
  author=soup.find('span', attrs={'class':'info_auth'}).find_all('a')

  author_list = [auth.get_text() for auth in author]
  authors = ','.join(author_list)
  return authors

```

```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup

gdown.download('https://bit.ly/3GisL6J','book.csv',quiet=False)
```

```python
def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']
  
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser')

  try:
    if pd.isna(title):
      title=soup.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass



  try:
    if pd.isna(author):
      author=soup.find('span', attrs={'class':'info_auth'}).find_all('a')
      author_list = [auth.get_text() for auth in author]
      author = ','.join(author_list)
  except AttributeError:
    pass




  try:
    if pd.isna(pub):
      pub = soup.find('span', attrs={'class':"authPub info_pub"}).find('a').get_text()
  except AttributeError:
    pass



  try:
    if pd.isna(year):
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  return title, author, pub, year


 
```
